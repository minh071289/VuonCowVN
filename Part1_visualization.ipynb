```python
df_priceHis.drop(df_priceHis.tail(7).index,inplace=True)
```

```python
df_priceHis['NGÀY'] = pd.to_datetime(df_priceHis['NGÀY'], format='%d/%m/%Y')
```

```python
df_priceHis = df_priceHis.sort_values('NGÀY')
df_priceHis.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))
plt.plot(df_priceHis['NGÀY'], df_priceHis['GIÁ ĐÓNG CỬA'], color='blue', marker='o', linewidth=1)

plt.title('Biểu đồ thể hiện biến động giá đóng cửa mỗi ngày 2021-2025')
plt.xlabel('Ngày')
plt.ylabel('Giá Đóng Cửa (VND)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Nhận xét:
- Biến động giá đóng cửa mỗi ngày từ năm 2021-2025 có xu hướng tăng rõ rệt
- Giai đoạn tháng 7/2023-7/2024 tăng mạnh mẽ nhất. Tháng 7/2024, giá cổ phiếu tăng hơn gấp đôi so với tháng 7/2023 .
- Nửa sau năm 2024, sau khi đạt đỉnh, giá cổ phiếu giảm nhẹ, nhưng sau đó duy trì cao ở mức ổn định. Và tiếp tục tăng cao vào cuối năm 2024, đầu 2025.


```python
# Lọc dữ liệu từ 01/07/2023 đến 01/07/2024
start_date = '2023-07-01'
end_date = '2024-07-01'
mask = (df_priceHis['NGÀY'] >= start_date) & (df_priceHis['NGÀY'] <= end_date)
filtered_df = df_priceHis.loc[mask]

plt.figure(figsize=(9, 5))
plt.plot(filtered_df['NGÀY'], filtered_df['GIÁ ĐÓNG CỬA'], marker='o', linewidth=1)

plt.title('Giá đóng cửa từ tháng 7/2023 đến tháng 7/2024')
plt.xlabel('Ngày')
plt.ylabel('Giá đóng cửa')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Nhận xét: Trong vòng một năm, từ 7/2023-7/2024 giá cổ phiếu tăng gấp đôi. Cuối tháng 4/2024 là thời điểm có sự tăng mạnh mẽ và liên tục.

```python
df_priceHis['quarter'] = df_priceHis['NGÀY'].dt.to_period('Q')
# Gộp theo quý tính trung bình
df_quarter = df_priceHis.groupby('quarter')['GIÁ ĐÓNG CỬA'].mean().reset_index()

plt.figure(figsize=(9,5))

plt.bar(df_quarter['quarter'].astype(str), df_quarter['GIÁ ĐÓNG CỬA'], width=0.4)
plt.title('Biểu đồ thể hiện trung bình giá đóng cửa theo từng quý các năm')
plt.xlabel('Quý')
plt.ylabel('Trung bình giá đóng cửa')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Nhận xét:
- Những năm 2021 đến đầu năm 2023, giá cổ phiểu các quý ở mức ổn định. Từ nửa sau năm 2023 giá tăng nhanh và có xu hướng tiếp tục tăng trong tương lai.
- Xu hướng tăng dài hạn thể hiện tâm lý thị trường khá tích cực và có thể do các yếu tố hỗ trợ về kinh tế, chính sách hoặc kết quả kinh doanh tốt của các công ty.

```python
plt.figure(figsize=(10,5))
plt.bar(df_priceHis['NGÀY'], df_priceHis['KHỐI LƯỢNG'])
plt.xlabel('Ngày')
plt.ylabel('Khối lượng giao dịch')
plt.title('Biểu đồ thể hiện khối lượng giao dịch mỗi ngày')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

Nhận xét: Khối lượng giao dịch có những khoảng tăng vọt, rất cao so với khối lượng trung bình.


```python
df_priceHis['MA_20'] = df_priceHis['KHỐI LƯỢNG'].rolling(window=20).mean()
df_priceHis['MA_50'] = df_priceHis['KHỐI LƯỢNG'].rolling(window=50).mean()

plt.figure(figsize=(10, 5))
plt.plot(df_priceHis['NGÀY'], df_priceHis['KHỐI LƯỢNG'], label='Khối lượng giao dịch', alpha=0.4)
plt.plot(df_priceHis['NGÀY'], df_priceHis['MA_20'], label='Moving Average 20 ngày', color='red', linewidth=2)
plt.plot(df_priceHis['NGÀY'], df_priceHis['MA_50'], label='Moving Average 50 ngày', color='green', linewidth=2)

plt.title('Khối lượng giao dịch và Moving Average')
plt.xlabel('Ngày')
plt.ylabel('Khối lượng giao dịch')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Nhận xét:
- Khối lượng giao dịch có những khoảng tăng vọt, có nhiều khoảng tăng, giảm rõ rệt
- Năm 2024: Có một đợt khối lượng tăng rất mạnh, vượt xa mức bình thường, kèm theo MA 20 cắt lên MA 50 thể hiện dấu hiệu thị trường sôi động.
- Giai đoạn 2022-2023: Khối lượng giao dịch duy trì ở mức thấp, cả hai đường MA đi ngang hoặc giảm dần, cho thấy thị trường trầm lắng.
- Cuối 2024 - đầu 2025: Cả hai đường MA đang có dấu hiệu đi lên lại, có thể cho thấy sự phục hồi.
